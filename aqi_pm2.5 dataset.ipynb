```python
import pandas as pd
import numpy as np
df = pd.read_csv("./datafile/LA_pm2.5.csv")
df.head()
```

```python
df = df.drop(columns = ["Policy Area", "Dataset", "Variable","Row ID","Tract Number"])
df.head()
```

```python
months = [None, "January", "February", "March", "April", "May", 
          "June", "July", "August", "September", "October", "November", "December" ]
month_df = []
date_track = ""
for i in range(len(df)):
    for j in range(len(df["Date"][i])):
        if df["Date"][i][j] != "/":
            date_track += df["Date"][i][j]
        else:
            month_int = int(date_track)
            month_df.append(months[month_int])
            date_track = ""
            break

df["Month"] = month_df
df.head()
```

```python
df["Date"].value_counts()
```

```python
df = df.drop(columns = "Date")
df.head()
```

```python
df = df.drop(columns = "GEOID")
df.head()
```

```python
def create_geography(geoPoints):
    longitude = []
    latitude = []
    forbidden = [",", "(", ")"]
    split_str = []
    for i in range(len(geoPoints)):
        split_str.append(geoPoints[i].replace(")", "").replace("(", "").replace(" ", "").split(","))
    for j in range(len(split_str)):
        longitude += [split_str[j][0]]
        latitude += [split_str[j][1]]
    return [longitude, latitude]

geoPoints = []
for i in range(len(df["Location"])):
    geoPoints.append(df["Location"][i])

newCoordinates = create_geography(geoPoints)
```

```python
df["Longitude"] = newCoordinates[0]
df.head()
```

```python
df["Latitude"] = newCoordinates[1]
df.head()
```

```python
import reverse_geocoder as rg 

def reverseGeocode(coordinates): 
    result = rg.search(coordinates)
    return (result)


coordinates =list(zip(df['Longitude'],df['Latitude'])) # generates pair of (lat,long)
data = reverseGeocode(coordinates)

df['CITY'] = [i["name"] for i in data]

df.head()
```
