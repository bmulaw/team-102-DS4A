```python
# SUMMARY:

"""

    This is a quick analysis on our merged dataset about the AQI and HQTA. The merged
    dataset was messy so most of the code is cleaning it. 

"""
```

```python
import pandas as pd
df = pd.read_csv("./datafile/merged_AQI_HQTA.csv")
df.head()
```

```python
df.columns
```

```python
df.head()
```

```python
# total sum of non-Hispanic population by each 4 cities, 
# overwrites the original 

a = df.groupby("City")["TotalNotHi"].sum()
nonHis = []
for i in df["City"]:
    nonHis.append(a[i])


df["TotalNotHispanic"] = nonHis
df.head()
```

```python
# total sum of White population by each 4 cities, 
# overwrites the original White column in dataset

w = df.groupby("City")['White'].sum()
wList = []
for i in df["City"]:
    wList.append(w[i])


df["White"] = wList
df.head()
```

```python
# total sum of Black population by each 4 cities, 
# overwrites the original Black column in dataset

b = df.groupby("City")['Black'].sum()
bList = []
for i in df["City"]:
    bList.append(b[i])


df['Black'] = bList
df.head()
```

```python
# total sum of Native population by each 4 cities, 
# overwrites the original Native column in dataset

n = df.groupby("City")['Native'].sum()
nList = []
for i in df["City"]:
    nList.append(n[i])


df['Native'] = nList
df.head()
```

```python
# total sum of Asian population by each 4 cities, 
# overwrites the original Asian column in dataset

asian = df.groupby("City")['Asian'].sum()
aList = []
for i in df["City"]:
    aList.append(asian[i])


df['Asian'] = aList
df.head()
```

```python
# total sum of Native Pacific population by each 4 cities, 
# overwrites the original Native Pacific column in dataset

np = df.groupby("City")['NativePaci'].sum()
npList = []
for i in df["City"]:
    npList.append(np[i])


df['NativePaci'] = npList
df.head()
```

```python
# total sum of Hispanic population by each 4 cities, 
# overwrites the original Hispanic column in dataset

h = df.groupby("City")['Hispanic'].sum()
hList = []
for i in df["City"]:
    hList.append(h[i])


df['Hispanic'] = hList
df.head()
```

```python
# total sum of the overall population of each 4 cities, 
# overwrites the original Total column in dataset

total = df.groupby("City")["Total"].sum()
totL = []
for i in df["City"]:
    totL.append(total[i])


df["Total"] = totL
df.head()
```

```python
# total sum of other non-Hispanic population by each 4 cities, 
# overwrites the original OtherNotHi column in dataset 

onh = df.groupby("City")["OtherNotHi"].sum()
onhList = []
for i in df["City"]:
    onhList.append(onh[i])


df["OtherNotHi"] = onhList
df.head()
```

```python
# Dropped unncessary columns (like Year because all the data are from 2015)

df = df.drop(columns = ["Site ID", "Year", "BLKGRPCE", "GEOID", "NAMELSAD", "ACRES"])
df.head()
```

```python
# Renamed the "New Date" column to "Date" - it's cleaner that way

df = df.rename(columns={'New Date': 'Date'})
df.head()
```

```python
# check the max PM10 value

max(df["PM10 AQI Value"])
```

```python
# check the min PM10 value

min(df["PM10 AQI Value"])
```

```python
df["PM10 AQI Value"].mean()
```

```python
# making a new dataframe with AQI lower than the mean

low_pm10 = df[df["PM10 AQI Value"] >= 36]
low_pm10
```

```python
# making a new dataframe with the lowest AQI

lowest_pm10 = df[df["PM10 AQI Value"] == 15]
lowest_pm10.head()
```

```python
# a quick overview of how frequently each AQI measure show up in the dataset

df["PM10 AQI Value"].value_counts()
```
