```python
hqta_df = hqta_df.drop(columns= ["OBJECTID_1", "FID_Census", "OBJECTID", "STATEFP","COUNTYFP", 
                                 "TRACTCE", "MTFCC", "FUNCSTAT", "Shape_Leng","ID2", "NumGEOID",
                                 "Shapearea", "Shapelen", "Shape_Le_1", "Shape_Area", "NAME",
                                 "AWATER", "FID_High_Q", "OBJECTID_2", "GEOid_1",
                                 "GEOid2", "GEOid3", "GEOIDTEXT"])
hqta_df.columns
```

```python
import reverse_geocoder as rg 

def reverseGeocode(coordinates): 
    result = rg.search(coordinates)
    return (result)


coordinates =list(zip(hqta_df['INTPTLAT'],hqta_df['INTPTLON'])) # generates pair of (lat,long)
data = reverseGeocode(coordinates)

hqta_df['CITY'] = [i["name"] for i in data]

hqta_df.head()
```

```python
len(hqta_df.columns)
```

```python
hqta_df["NAMELSAD"].unique()
```

```python
hqta_df["NAMELSAD"] = [hqta_df["NAMELSAD"][i][-1] for i in range(len(hqta_df))]
hqta_df["NAMELSAD"].head(5)
```

```python
hqta_df = hqta_df.replace("MULTIPOLYGON", "", regex=True)
hqta_df["the_geom"][0]
```
